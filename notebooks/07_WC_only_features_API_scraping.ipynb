```python

import http.client
import json
import time
import pandas as pd

API_KEY = "0c9c8577f160283e4884c83dac66c727"
HOST    = "v3.football.api-sports.io"
HEADERS = {"x-apisports-key": API_KEY}

SEASONS = [2006, 2010, 2014, 2018, 2022]

STAT_COLS = [
    "Shots on Goal", "Shots off Goal", "Total Shots", "Blocked Shots",
    "Shots insidebox", "Shots outsidebox", "Corner Kicks", "Offsides",
    "Ball Possession", "Goalkeeper Saves", "Total passes",
    "Passes accurate", "Passes %",
]

BASE_COLS = [
    "fixture.id", "fixture.date", "fixture.status.elapsed",
    "league.id", "league.name", "league.season", "league.round",
    "teams.home.id", "teams.home.name", "teams.away.id", "teams.away.name",
    "goals.home", "goals.away",
    "score.halftime.home", "score.halftime.away",
    "score.fulltime.home", "score.fulltime.away",
    "score.extratime.home", "score.extratime.away",
]


def api_get(path):
    """Open a fresh connection per request to avoid stale socket errors."""
    conn = http.client.HTTPSConnection(HOST)
    conn.request("GET", path, headers=HEADERS)
    res = conn.getresponse()
    raw = res.read()
    conn.close()
    return json.loads(raw)


all_rows = []

for season in SEASONS:
    print(f"\n=== Season {season} ===")

    # --- 1. Fetch all fixtures for the season ---
    fixtures_data = api_get(f"/fixtures?league=1&season={season}")
    time.sleep(0.5)

    fixtures = fixtures_data.get("response", [])
    df_season = pd.json_normalize(fixtures)

    # Keep only columns that exist in this season's data
    present = [c for c in BASE_COLS if c in df_season.columns]
    df_season = df_season[present]

    print(f"  {len(df_season)} fixtures found")

    for _, row in df_season.iterrows():
        fixture_id = int(row["fixture.id"])
        home_id    = int(row["teams.home.id"])
        away_id    = int(row["teams.away.id"])

        record = row.to_dict()

        # --- 2. Statistics ---
        for col in STAT_COLS:
            record[f"home_{col}"] = None
            record[f"away_{col}"] = None

        try:
            stat_data = api_get(f"/fixtures/statistics?fixture={fixture_id}")
            time.sleep(0.35)
            for team_block in stat_data.get("response", []):
                tid    = team_block["team"]["id"]
                stats  = {s["type"]: s["value"] for s in team_block["statistics"]}
                prefix = "home" if tid == home_id else "away"
                for col in STAT_COLS:
                    record[f"{prefix}_{col}"] = stats.get(col)
        except Exception as e:
            print(f"    [stats error] fixture {fixture_id}: {e}")

        # --- 3. Lineups (formation only) ---
        record["formation_home"] = None
        record["formation_away"] = None

        try:
            lineup_data = api_get(f"/fixtures/lineups?fixture={fixture_id}")
            time.sleep(0.35)
            for team_block in lineup_data.get("response", []):
                tid       = team_block["team"]["id"]
                formation = team_block.get("formation")
                if tid == home_id:
                    record["formation_home"] = formation
                else:
                    record["formation_away"] = formation
        except Exception as e:
            print(f"    [lineup error] fixture {fixture_id}: {e}")

        all_rows.append(record)
        print(f"  fixture {fixture_id} done", end="\r")

    print(f"  Season {season} complete ({len(df_season)} fixtures)")

df_all_wc = pd.DataFrame(all_rows).rename(columns={"fixture.id": "fixture_id"})
print(f"\nTotal rows: {df_all_wc.shape[0]}, columns: {df_all_wc.shape[1]}")
df_all_wc.head()
```
